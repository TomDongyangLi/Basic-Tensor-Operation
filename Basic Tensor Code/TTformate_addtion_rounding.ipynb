```python
import torch
from Utils import data_generator,TT_add,TT_to_tensor,TT_rounding,TT_rounding_2,TTSVD
```

```python
d = 3                   # tensor orde
I = 5                  # tensor dimension
r = 2                   # choose tt-rank to be constant
                        # need rank to be smaller than I/2, otherwise the mu-orthgonalization will not work
rank = [r] * (d - 1)    # tt-rank
shape = [I] * d 
```

```python
core1 = data_generator(shape, rank)
core2 = data_generator(shape, rank)
```

```python
core_add = TT_add(core1,core2)
print(core_add)
```

```python
X_1 = TT_to_tensor(rank,core1)
X_2 = TT_to_tensor(rank,core2)
X_add = TT_to_tensor([2*r] * (d - 1),core_add)
```

```python
print(X_1 + X_2)
```

```python
print(X_add)
```

```python
print(torch.norm(X_1 + X_2 - X_add)/torch.norm(X_1 + X_2))
```

```python
core_round = TT_rounding(core_add,r)
core_round_2 = TT_rounding_2(core_add,r)
```

```python
for i in range(d):
    print(core_round[i].shape)
    print(core_round_2[i].shape)
```

```python
X_round = TT_to_tensor(rank,core_round)
X_round_2 = TT_to_tensor(rank,core_round_2)
```

```python
print(X_round)
```

```python
print(X_round_2)
```

```python
print(torch.norm(X_round - X_add)/torch.norm(X_add))
print(torch.norm(X_round_2 - X_add)/torch.norm(X_add))
```

#### We have error, this makes sense since TT-rounding is a way of projecting the tensor into the manifold with fixed rank
#### And below we can see that rounding with SVD from left to right is the same as TTSVD
#### But TTSVD does not require orthogonalization from right to left as prestep, GPT says that this step makes sure that SVD can be computed more stable

```python
X_add_proj = TT_to_tensor(rank,TTSVD(X_add,rank))
print(torch.norm(X_add_proj - X_add)/torch.norm(X_add))
print(torch.norm(X_add_proj - X_round)/torch.norm(X_round))
```

```python
print(X_add_proj)
```

#### However, we still have difference between "orthogonalization from left to right and then SVD from right to left" and "orthogonalization from right to left and then SVD from left to right"

Maybe it's because we are truncating the SVD for different matrices/rank in each case. 
TTSVD and the second case are the same since they are truncating the same group of matrices/rank
